```python
# 1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a three-
# dimensional space
import numpy as np
import pandas as pd
import plotly.express as px
np.random.seed(30)
data = {
'X': np.random.uniform(-10, 10, 300),
'Y': np.random.uniform(-10, 10, 300),
'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)
df
fig=px.scatter_3d(x=df['X'],y=df['Y'],z=df['Z'])
fig.show()
```

```python
# 2. Using the Student Grades, create a violin plot to display the distribution of scores across different grade
# categories.
np.random.seed(15)
data = {
'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)
fig = px.violin(df, x="Grade",y='Score')
fig.show()
```

```python
# 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
# days.

import numpy as np
import pandas as pd
import plotly.express as px

# Sample data
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)
heatmap_data = df.groupby(['Month', 'Day'])['Sales'].sum().unstack()

# Create the heatmap using Plotly
fig = px.imshow(heatmap_data, labels=dict(x="Day of Month", y="Month", color="Sales"))
fig.update_layout(
    title='Sales Variation Across Months and Days',
)
fig.show()
```

```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function
import numpy as np
import pandas as pd
import plotly.express as px
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
'X': x.flatten(),
'Y': y.flatten(),
'Z': z.flatten()
}
df = pd.DataFrame(data)
fig =px.scatter_3d(x=df['X'],y=df['Y'],z=df['Z'])
fig.show()
# df
```

```python
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
import numpy as np
x=np.linspace(0,11,100)
y=np.sin(x)
from bokeh.plotting import figure,output_file,show
output_file('Q1.html')
p=figure(title = "Sin wave ")
p.xaxis.axis_label='x-axis'
p.yaxis.axis_label='y-axis'
p.line(x,y)
show(p)

```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
np.random.seed(40)
x = np.random.rand(20)
y = np.random.rand(20)
sizes = np.random.randint(5, 20, 20)
colors = ["red", "blue", "green", "purple", "orange"]
p = figure(width=600, height=400)

for i in range(len(x)):
    p.circle(x[i], y[i], size=sizes[i], color=colors[i % len(colors)], alpha=0.6)

# Show the plot
output_notebook()
show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
p = figure(x_range=fruits, height=350, title="Fruit Counts")

p.vbar(x=fruits, top=counts, width=0.9)
show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.
np.random.seed(20)
data_hist = np.random.randn(1000)
# print(data_hist)
hist, edges = np.histogram(data_hist, bins=30)
p=figure(width=600,height=400)
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
         fill_color="skyblue", line_color="white",)
show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import linear_cmap
import numpy as np

# Provided dataset
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

p = figure(title='Heatmap Example')

# Add the heatmap
p.rect(x='x', y='y', width=1, height=1, source={'x': xx.flatten(), 'y': yy.flatten(), 'value': data_heatmap.ravel()},
       fill_color=linear_cmap('value', 'Viridis256', data_heatmap.min(), data_heatmap.max()))


# Show the plot
show(p)

```
